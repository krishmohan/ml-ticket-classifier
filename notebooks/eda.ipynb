```python
import pandas as pd

train = pd.read_json("../data/splits/train.jsonl", lines=True)

train.head(), train.info()
```

```python
train["label"].value_counts().plot(kind="bar")
```

```python
train["length"] = train["text"].apply(lambda x: len(x.split()))

train["length"].describe()
```

```python
for i in range(5):
    print(train.iloc[i]["text"])
    print("Label:", train.iloc[i]["label"])
    print("-" * 50)
```

```python
vocab = set()

for text in train["text"]:
    vocab.update(text.lower().split())

len(vocab)
```
